```python
categories = ['bpd', 'anxiety', 'depression', 'bipolar', 'mentalillness',
       'schizophrenia', 'ocd', 'normal', 'ptsd', 'suicidal', 'offmychest',
       'stress', 'personality disorder', 'adhd', 'bdd', 'eating disorder',
       'hoarding disorder', 'panic_disorder', 'social anxiety']
target_size = 7500
balanced_df_list = []
for category in categories:
    category_df = merged_df[merged_df['label'] == category]
    if len(category_df) >= target_size:
        balanced_df_list.append(category_df.sample(n=target_size, random_state=42).reset_index(drop=True))
    else:
        balanced_df_list.append(category_df)
merged_df1 = pd.concat(balanced_df_list).reset_index(drop=True)
merged_df1 = merged_df1.sample(frac=1, random_state=42).reset_index(drop=True)
```
